```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import glob
```

```python
file_list = glob.glob("multi_cluster_res/*.csv")
plt.style.use("seaborn-v0_8")
palette = sns.color_palette("tab10")
for file in file_list:
    for seed in range(0,5):
        if "action_come" in file:
            df = pd.read_csv(file,index_col=0)
            plot_df = df[(df['seed'] == seed)& (df['topk_cluster'] <= 13)]
            plt.figure(figsize=(7,6))
            sns.lineplot(data=plot_df, x='topk_cluster', y='test_Y_NDCG_10_female', label='female', color = palette[3], marker='o')
            sns.lineplot(data=plot_df, x='topk_cluster', y='test_Y_NDCG_10_male', label='male', color = palette[0], marker='o')
            sns.lineplot(data=plot_df, x='topk_cluster', y='test_Y_NDCG_10', label='overall', color = palette[2], marker='o')
            file_name = file.split('/')[-1][:-4].split("_")
            # plt.title(rf'{file_name[0].capitalize()} $\rightarrow$ {file_name[1].capitalize()}', fontsize=13, fontweight='bold')
            plt.xticks([1, 3, 5, 7, 9, 11, 13])
            plt.xlabel('Number of Multi-interest',fontsize = 18, fontweight='bold')
            plt.ylabel('NDCG@10',fontsize = 18, fontweight='bold')
            plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.13), ncol=3, fontsize=20)
            plt.yticks(fontsize=14)
            plt.xticks(fontsize=14)
            plt.show()
```
