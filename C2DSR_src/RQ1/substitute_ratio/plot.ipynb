```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import glob
```

```python
file_list = glob.glob('./substitution_ratio_res/*')
file_list
```

```python
for file_name in file_list:
    # file_name= file_list[0]
    for seed in range(0,5):
        if "drama_sci" in file_name:
            plt.figure(figsize=(7, 6))
            plt.style.use("seaborn-v0_8")
            df = pd.read_csv(file_name,index_col=0)
            colors = sns.color_palette("tab10")
            mean_df= df[df["seed"]==seed].iloc[:]
            # mean_df = df.groupby('substitute_ratio')[["test_Y_NDCG_10","test_Y_NDCG_10_male","test_Y_NDCG_10_female"]].mean()
            # mean_df = mean_df.iloc[:6]
            # print("seed:",seed)
            sns.lineplot(data=mean_df, x="substitute_ratio", y="test_Y_NDCG_10", label="Overall",color=colors[2],marker="o")
            sns.lineplot(data=mean_df, x="substitute_ratio", y="test_Y_NDCG_10_male", label="Male",color=colors[0],marker="o")
            sns.lineplot(data=mean_df, x="substitute_ratio", y="test_Y_NDCG_10_female", label="Female",color=colors[3],marker="o")
            file = file_name.split("/")[-1][:-4]
            file = file.split("_")
            # plt.title(rf'{file[0].capitalize()} $\rightarrow$ {file[1].capitalize()}', fontsize=16, fontweight='bold', x=0.5, y=1.05)
            plt.xlabel("Substitution ratio",fontsize=21)
            plt.ylabel("NDCG@10",fontsize=21)
            # plt.legend(fontsize=15, title_fontsize='x-large')
            plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.13), ncol=3, fontsize=20)
            plt.yticks(fontsize=14)
            plt.xticks(fontsize=14)
```

```python
for file_name in file_list:
    # file_name= file_list[0]
    plt.figure(figsize=(7, 4))
    plt.style.use("seaborn-v0_8")
    df = pd.read_csv(file_name,index_col=0)
    colors = sns.color_palette("tab10")
    mean_df = df.groupby('substitute_ratio')[["test_Y_NDCG_10","test_Y_NDCG_10_male","test_Y_NDCG_10_female"]].mean()
    # mean_df = mean_df.iloc[:6]
    sns.lineplot(data=mean_df, x="substitute_ratio", y="test_Y_NDCG_10", label="Overall",color=colors[2],marker="o")
    sns.lineplot(data=mean_df, x="substitute_ratio", y="test_Y_NDCG_10_male", label="Male",color=colors[0],marker="o")
    sns.lineplot(data=mean_df, x="substitute_ratio", y="test_Y_NDCG_10_female", label="Female",color=colors[3],marker="o")
    file = file_name.split("/")[-1][:-4]
    file = file.split("_")
    # plt.title(rf'{file[0].capitalize()} $\rightarrow$ {file[1].capitalize()}', fontsize=16, fontweight='bold', x=0.5, y=1.05)
    plt.xlabel("substitution ratio",fontsize=15, fontweight='bold')
    plt.ylabel("NDCG@10",fontsize=15, fontweight='bold')
```
